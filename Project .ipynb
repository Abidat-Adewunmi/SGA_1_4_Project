```python
#import functions
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline

sql = """
SELECT *
FROM attrition_records
"""

table = pd.read_sql(sql, conn)
table.head(10)
```

```python
table.describe()
```

```python
table["Attrition"].replace(["Yes","No"],[1,2],inplace=True)
table.head()
```

```python
table.drop(columns=["EmployeeCount","StandardHours"],axis = 1, inplace=True)
table.head()
```

```python
#representing correlation in heatmap
plt.figure(figsize=[25,10])
res = sns.heatmap(table.corr(), annot=True, fmt = ".3f")
plt.title ("Correlation of columns in table", fontsize = 30);
```

```python
#representing correlation in tabular format
table.corr()
```

```python
# 1. The three (3) factors behind employee attrition are : 
#TotalWorkingYears - 0.171, JobLevel- 0.169, and YearsInCurrentRole - 0.161
```

```python
#2. the relationship between monthly income and Education - 0.095

plt.figure(figsize=[4,2])
sns.heatmap(table[['Education','MonthlyIncome']].corr(), annot=True, fmt = ".3f")
plt.title ("Correlation of Education and Monthly Income");
```

```python
#3. Age has strong effect on the attrition - 0.159
plt.figure(figsize=[4,2])
sns.heatmap(table[['Age','Attrition']].corr(), annot=True, fmt = ".3f")
plt.title ("Effect of Age on Attrition");
```

```python
#4. Income is not the main factor in Employee attrition 
plt.figure(figsize=[4,2])
sns.heatmap(table[['MonthlyIncome','Attrition']].corr(), annot=True, fmt = ".3f")
plt.title ("Correlation of Monthly Income and Attrition");
```

```python
#5. Worklife balance have little impact on overall attrition rate - 0.064
plt.figure(figsize=[4,2])
sns.heatmap(table[['WorkLifeBalance','Attrition']].corr(), annot=True, fmt = ".3f")
plt.title ("Impact of Worklife balance on Attrition rate");
```
